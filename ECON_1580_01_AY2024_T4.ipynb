```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
conventaional_gas = pd.read_csv("./New_York_Harbor_Conventional_Gasoline_Regular_Spot_Price_FOB.csv")
wti = pd.read_csv("./Cushing_OK_WTI_Spot_Price_FOB.csv")
```

```python
percent_change_wti = wti["Cushing OK WTI Spot Price FOB Dollars per Barrel"].pct_change()
pct_conventaional_gas = conventaional_gas["New York Harbor Conventional Gasoline Regular Spot Price FOB Dollars per Gallon"].pct_change()
```

```python
plt.plot(conventaional_gas["Year"],
         pct_conventaional_gas,
         label="conventional gas")
plt.plot(wti["Year"],
         percent_change_wti,
         label="crude oil", c="r", alpha=0.5)
plt.legend()
plt.xlabel("year")
plt.ylabel("price change")
plt.show()
```
